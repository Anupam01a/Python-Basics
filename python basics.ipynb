1. What is Python, and why is it popular ?

 - Python is a high-level, open-source programming language known for its simplicity and readability. It’s popular because it’s easy to learn, versatile, and widely used across different fields like web development, data science, and automation. It also has a huge collection of libraries and frameworks that make development faster and easier.

2. What is an interpreter in Python ?

  - Python is an interpreted language, which means its code runs line by line. When we execute a Python program, the interpreter first converts the code into bytecode and then runs it.


3. What are pre-defined keywords in Python ??

  - Keywords in Python are predefined reserved words that have special meanings and specific purposes.
They’re used to define the structure and rules of Python programs, and we can’t use them as variable or function names.
Example: if, while, class, def, return.

4. Can keywords be used as variable names ?

 - No, keywords cannot be used as variable names in Python.
They are reserved words that have predefined meanings and are used to define the syntax and structure of the language.
Using a keyword as a variable name will result in a SyntaxError.

5. What is mutability in Python ??

 - Mutability in Python refers to the ability of an object to change its state or content after it has been created.
Objects whose values can be modified are called mutable objects, while those that cannot be changed are immutable objects.



 6. Why are lists mutable, but tuples are immutable ?

  - Lists are mutable because their elements can be changed, added, or removed after creation.
 - Tuples are immutable because their values are fixed once created.

7. What is the difference between “==” and “is” operators in Python ?

 - In Python, == compares the values of two objects to check if they are equal, while is checks whether both variables refer to the same object in memory.

8. What are logical operators in Python ?

 - Logical operators in Python are used to combine conditional statements and return True or False based on the conditions.

There are three logical operators in Python:

 - and → Returns True if both conditions are true.
Example: a > 10 and b < 5

 - or → Returns True if at least one condition is true.
Example: a > 10 or b < 5

 - not → Reverses the result; returns True if the condition is false.
Example: not(a > 10)

9. What is type casting in Python ??

 - Type casting in Python means converting one data type into another, such as converting an integer to a string or a float.
It’s done using functions like int(), float(), str(), etc.

10. What is the difference between implicit and explicit type casting

 - Implicit type casting (automatic):
Python automatically converts one data type to another during an operation, without the programmer’s involvement is called as implicit type casting.
 - Explicit type casting (manual):
The programmer manually converts one data type to another using functions like int(), float(), or str() is called as explicit type casting.


11. What is the purpose of conditional statements in Python ?

 - The purpose of conditional statements in Python is to make decisions in a program based on certain conditions.They allow the program to execute different blocks of code depending on whether a condition is True or False.

12. How does the elif statement work ?

 - The elif statement in Python stands for “else if” and is used to check multiple conditions one by one.
It runs only when the previous if condition is False, and its own condition is True.

```python
#For example:-
x = 10
if x > 15:
    print("Greater than 15")
elif x > 5:
    print("Greater than 5 but not 15")
else:
    print("5 or less")
```

13. What is the difference between for and while loops ?

 - The for loop is used when you know in advance how many times you want to repeat a block of code — it iterates over a sequence

 - The while loop is used when you want to repeat code until a certain condition becomes false.

```python
#for example

# for loop
for i in range(5):
    print(i)

# while loop
i = 0
while i < 5:
    print(i)
    i += 1
```

14. Describe a scenario where a while loop is more suitable than a for loop?

 - A while loop is more suitable when the number of iterations is not known in advance and depends on a condition being met during runtime.


```python
#Example

password = ""
while password != "assignment":
    password = input("Enter password: ")
print("Access granted!")

```

```python
# 6.  Write a function to demonstrate the behavior of mutable and immutable arguments

def change_values(imm, mut):

    imm = imm + 5        # This creates a new integer object
    mut.append("new")    # This changes the original list

    print("Inside function:")
    print("  imm =", imm)  # new value
    print("  mut =", mut)  # changed list

# Outside:
x = 10        # immutable
y = [1, 2, 3] # mutable

print("Before:", x, y)
change_values(x, y)
print("After:", x, y)
```

```python
# 8. Write a program to demonstrate the use of logical operators.

a = 10
b = 20
c = 30

# AND operator
print(f"a > b and a < c: {a > b and a < c}") # True and True = True
print(f"a > c and a < b: {a > c and a < b}") # False and False = False

# OR operator
print(f"a > b or a > c: {a > b or a > c}")   # True or False = True
print(f"b > a or c < a: {b > a or c < a}")   # False or False = False

# NOT operator
print(f"not(a > b): {not(a > b)}")           # not True = False
print(f"not(b > a): {not(b > a)}")
```

```python
# 14. Write a program to reverse a string using a while loop?

string = input("enter a string")
rev = ""
i = -1

while i >= -len(string):
  rev = rev + string[i]

  i = i -1

  print("reversed string :", rev)
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop

num = int(input("Enter a number: "))

fact=1
i = 1

while i<=num:
  fact = fact * i
  i = i + 1

  print("factorial is:", fact)

```
